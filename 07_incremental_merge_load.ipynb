```sql
CREATE OR REPLACE TABLE nyc_taxi_silver_incremental AS
SELECT
    *,
    current_timestamp() AS last_updated
FROM nyc_taxi_silver
LIMIT 1000;
```

```sql
CREATE OR REPLACE TABLE nyc_taxi_stage AS

SELECT
    *,
    current_timestamp() AS last_updated
FROM nyc_taxi_silver
LIMIT 1500;
```

```sql
SELECT * FROM nyc_taxi_silver_incremental LIMIT 1;
SELECT * FROM nyc_taxi_stage  LIMIT 1
```

```sql
MERGE INTO nyc_taxi_silver_incremental AS target

USING (
SELECT *
FROM nyc_taxi_stage
) source

ON
AND target.pickup_datetime = source.pickup_datetime

WHEN MATCHED THEN
UPDATE SET
target.fare_amount = source.fare_amount,
target.trip_distance = source.trip_distance,
target.last_updated = current_timestamp()

WHEN NOT MATCHED THEN
INSERT *;
```
