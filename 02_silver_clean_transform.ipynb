Execute Deduplication on Result Table.

```sql
CREATE OR REPLACE TABLE nyc_taxi_silver AS
SELECT
  *
FROM
  (
    SELECT
      *,
      ROW_NUMBER() OVER (
          PARTITION BY
            tpep_pickup_datetime,
            tpep_dropoff_datetime,
            trip_distance,
            fare_amount,
            pickup_zip,
            dropoff_zip
          ORDER BY ingestion_time DESC
        ) AS rn
    FROM
      nyc_taxi_bronze
  ) t
WHERE
  rn = 1
  AND fare_amount > 0
  AND trip_distance > 0;
```
